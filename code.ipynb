The data were collected to test the water quality of the Bellinger River and Kalang River to help protect the snapping turtle's habitat and improve the river's ecosystem. Some community scientists and volunteers collected the data at multiple sites. These data record changes in water quality between 2017 and 2021. Water quality can be assessed by testing seven variables - air temperature, water temperature, pH, electrical conductivity, available phosphate, dissolved oxygen and turbidity. The data will be uploaded and stored in the NSW Waterwatch database.

### 3.2 Outliers

Generally speaking, we can first find the outliers of variables by observing the maximum and minimum of each variable. Then, the extreme value is compared with the mean, standard deviation and quartile to determine who is more inclined to the outliers. For example, for electrical conductivity, the maximum is 51500, the minimum is 7, the mean is 3286, and 75% of the value is below 100. Therefore, we can know that the maximum is too large, and its nearby values are outliers.

Outliers can also be found through a boxplot. For example, the following figure shows that the outliers of pH units are about below 5.8 and above 7.8.

# 4 Exploratory analysis

```python
data_variables = pd.read_excel('bellingen-riverwatch-ongoing-data_last-updated-20210812_ig.xlsx')
#extract core variables
cols = ['SITE CODE','Survey Date','Survey duration (hrs)','Water type','Water level or flow','Air Temperature (deg C)','Water Temperature (deg C)','pH (pH units)','Electrical Conductivity (µS/cm)','Turbidity (NTUs)','Available Phosphate (mg/L)','Dissolved Oxygen (mg/L)','Dissolved Oxygen (% sat)','Faecal Coliforms (CFU/100mL)','Nitrates (mg/L as N)']
new_data = data[cols]
```

## 4.1 Univariate Non-Graphical

In this section, I have used the six rating criterias for water quality to rate the rivers. I did not rank specific rivers or sites because I wanted to get an overall picture of these rivers.

```python
ratings_A = []
for i in new_data['pH (pH units)']:

    if i >= 6 and i <= 8:
        ratings_A.append('Good')
    elif i < 6 or i > 8:
        ratings_A.append('Poor')
    else:
        ratings_A.append('Unknown')
#reference:https://www.geeksforgeeks.org/create-a-column-using-for-loop-in-pandas-dataframe/
        
new_data['pH ratings'] = ratings_A
        
new_data[['pH (pH units)','pH ratings']]
```

```python
print(new_data['pH ratings'].value_counts())
```

Based on the pH ratings and counts, the water quality of these rivers was good at most of the sites and time.

```python
ratings_B = []
for i in new_data['Electrical Conductivity (µS/cm)']:

    if i < 300:
        ratings_B.append('Very Good')
    elif i >= 300 and i <= 500:
        ratings_B.append('Good')
    elif i > 500 and i <= 800:
        ratings_B.append('Fair')
    elif i > 800:
        ratings_B.append('Poor')
    else:
        ratings_B.append('Unknown')

        
new_data['Conductivity ratings'] = ratings_B
        
new_data[['Electrical Conductivity (µS/cm)','Conductivity ratings']]
```

```python
print(new_data['Conductivity ratings'].value_counts())
```

All five rivers had very good water quality in terms of conductivity classes.

```python
ratings_C = []

for i in new_data['Turbidity (NTUs)']:
    if i < 7:
         ratings_C.append('Very Good')
    elif i >= 7 and i < 10:
        ratings_C.append('Good')
    elif i >= 10 and i <= 30:
        ratings_C.append('Fair')
    elif i > 30:
        ratings_C.append('Poor')
    else:
        ratings_C.append('Unknown')
        
new_data['Turbidity ratings'] = ratings_C
        
new_data[['Turbidity (NTUs)','Turbidity ratings']]
```

```python
print(new_data['Turbidity ratings'].value_counts())
```

In terms of water turbidity ratings, most of the water quality is good. However, the counts indicate that the water quality was dair at some sites or periods.

```python
ratings_D = []

for i in new_data['Available Phosphate (mg/L)']:
    if i < 0.02:
         ratings_D.append('Very Good')
    elif i >= 0.02 and i < 0.05:
        ratings_D.append('Good')
    elif i >= 0.05 and i <= 0.15:
        ratings_D.append('Fair')
    elif i > 0.15:
        ratings_D.append('Poor')
    else:
        ratings_D.append('Unknown')
        
new_data['Available Phosphate ratings'] =  ratings_D    
new_data[['Available Phosphate (mg/L)','Available Phosphate ratings']]
```

```python
print(new_data['Available Phosphate ratings'].value_counts())
```

As for available phosphates, the amounts of these ratings are similar. However, we do not know which site or river has good or fair water quality.It is worth noting that the number of poor has increased compare to the previous variables.

```python
ratings_E = []

for i in new_data['Dissolved Oxygen (% sat)']:
    if i > 80 and i <= 90:
        ratings_E.append('Very Good')
    elif i > 90 and i < 100:
        ratings_E.append('Good')
    elif i >= 60 and i <= 80:
        ratings_E.append('Fair')
    elif i < 60 and i > 110:
        ratings_E.append('Poor')
    else:
        ratings_E.append('Unknown')
        
new_data['Dissolved Oxygen ratings'] = ratings_E  
new_data[['Dissolved Oxygen (% sat)','Dissolved Oxygen ratings']]
```

```python
print(new_data['Dissolved Oxygen ratings'].value_counts())
```

For dissolved oxygen, these ratings occupy almost the same proportion. 

```python
ratings_F = []

for i in new_data['Faecal Coliforms (CFU/100mL)']:
    if i == 0:
        ratings_F.append('Very Good')
    elif i < 150:
        ratings_F.append('Good')
    elif i < 1000:
        ratings_F.append('Fair')
    elif i > 1000:
        ratings_F.append('Poor')
    else:
        ratings_F.append('Unknown')

new_data['Faecal Coliforms ratings'] = ratings_F
new_data[['Faecal Coliforms (CFU/100mL)','Faecal Coliforms ratings']]
```

```python
print(new_data['Faecal Coliforms ratings'].value_counts())
```

This result indicates that most volunteers did not test the water quality with faecal coliforms. Although there was very little data to tell us that the water quality was good, the statistics were too small to make an accurate assessment.

```python
new_data.describe()
```

The ratings are also calculated by comparing the mean, standard deviation, and quartiles with the water quality guidelines. If the mean value falls into a specific rating range, then the water quality can be judged according to that criterion.

## 4.3 Multivariate Non-graphical

```python
#correlation
new_data.corr()
```

The closer the correlation value is to 1 or -1, the stronger the positive or negative linear correlation between the two variables. For example, the correlation between air temperature and water temperature is 0.816, indicating higher air temperature and higher water temperature. The correlation between water temperature and dissolved oxygen (mg/L) is -0.436 indicates that the higher the water temperature the lower the value of dissolved oxygen.

```python
#covariance
new_data.cov()
```

We consider only the sign when analysing the covariance. For example, pH is positively correlated with electrical conductivity and negatively correlated with turbidity.

```python
new_data.groupby(['Water Temperature (deg C)']).mean()[['Electrical Conductivity (µS/cm)']]
```

Group by () function tells us how water temperature relates to conductivity. For example, at 7 degrees Celsius, the mean of conductivity is 90.0 µS/cm.
